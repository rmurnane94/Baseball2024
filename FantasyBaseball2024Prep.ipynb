```python
import pandas
```

```python
ATC_projections = pandas.read_csv('ATC_projections.csv')
ATC_values = pandas.read_csv('ATC-auction-values.csv')
ATC_values.head()
```

```python
ATC_projections_simplified = ATC_projections.drop(columns = ['Name','G', 'AB', 'H', '1B', '2B', '3B', 'BB', 'SO', 'HBP', 'SF', 'SH', 'GDP', 'CS', 'BB%', 'IBB', 'K%', 'BB/K', 'ISO', 'BABIP', 'AVG', 'InterSD', 'InterSK', 'IntraSD', 'Spd', 'UBR', 'wSB', 'wRC', 'wRAA', 'BsR', 'Fld', 'Off', 'Def', 'WAR', 
                                                            'wRC+', 'Team', 'ADP'])
ATC_projections_simplified.head()
```

```python
ATC_values_simplified = ATC_values.drop(columns = ['Team', 'ADP', 'PA', 'mRBI', 'mR', 'mSB', 'mHR', 'mOBP', 'mSLG', 'PTS', 'aPOS'])
ATC_values_simplified.head()
```

```python
joined_ATC = ATC_values_simplified.merge(ATC_projections_simplified, left_on='PlayerId', right_on='PlayerId')
```

```python
final_ATC = joined_ATC[['Name', 'POS', 'Dollars', 'PA', 'HR', 'R', 'RBI', 'SB', 'OBP', 'SLG', 'OPS','wOBA']]
final_ATC.head()
```
